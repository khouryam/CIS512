## Create two new features:
Total Number of regular-fare tickets


```python
df['Total_Regular']=df['7 Day Unlimited']+df['30 Day Unlimited']+df['Full Fare']
df.head()
```

Ratio of Access-fare tickets to regular-fare tickets

```python
df['Ratio']= df['AccessPass']/ df['Total_Regular']
df.head()
```

# Create a linear regression visualization of full fares and access-related fares
Linear Regression - Regplot: How can this output be used to estimate the number of access-related fares if you have the number of full fares at the same station?


```python
sns.regplot(x='AccessPass',y='Full Fare', data=df)
```

This output can be used to estimate the number of access-related fares to the full fares at the same station by showing the comparison of the two fares. This graph potentially shows where the number of access fares are going and the number of full fares.
